```python
#!/bin/env python3
import os
import pathlib
import shutil
import tqdm.notebook as tqdm
```

```python
subjects = list(pathlib.Path("").glob("*.xls"))
boilers = [
    pathlib.Path("/.xls"),
    pathlib.Path("/.xls"),
]
input_folder = pathlib.Path("input")
for x in tqdm.tqdm(subjects):
    if x.stem in ["", ""]:
        continue
    shutil.rmtree(input_folder, ignore_errors=True)
    input_folder.mkdir()
    shutil.copy(x, input_folder)
    for y in boilers:
        shutil.copy(y, input_folder)
    assert os.system("./main.py") == 0
    assert os.system("typst compile main.typ") == 0
    os.chdir("tex-contract")
    assert os.system("pdflatex a4.tex") == 0
    pathlib.Path("a4.pdf").replace(f"../releases/{x.stem}.pdf")
    os.chdir("..")
```
